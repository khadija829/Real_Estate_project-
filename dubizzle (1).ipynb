```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd

# Base URL and config
BASE_URL = "https://www.dubizzle.com.om"
URL_TEMPLATE = BASE_URL + "/en/properties/properties-for-sale/?page={}"
MAX_PAGES = 119

# Data storage
dubizzle_data = {
    'property_title': [],
    'location': [],
    'number_of_rooms': [],
    'price': [],
    'size': [],
}

current_page = 1

while True:
    url = URL_TEMPLATE.format(current_page)
    print(f"Fetching page {current_page}: {url}")

    try:
        response = requests.get(url)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"❌ Failed to fetch page {current_page}: {e}")
        break

    soup = BeautifulSoup(response.content, 'html.parser')
    listings = soup.find_all('li', attrs={"aria-label": "Listing"})

    if not listings:
        print(f"❌ No listings found on page {current_page}. Stopping.")
        break

    print(f"✅ {len(listings)} listings found on page {current_page}")

    for listing in listings:
        # Property title
        title_tag = listing.find('h2', class_='_562a2db2')
        title = title_tag.text.strip() if title_tag else "N/A"
        dubizzle_data['property_title'].append(title)

        # Location
        loc_tag = listing.find('span', class_='f7d5e47e')
        location = loc_tag.text.strip() if loc_tag else "N/A"
        dubizzle_data['location'].append(location)

        # Number of rooms
        bed_tag = listing.find('span', attrs={'aria-label': 'Beds'})
        rooms_span = bed_tag.find('span', class_='_3e1113f0') if bed_tag else None
        rooms = rooms_span.text.strip() if rooms_span else "N/A"
        dubizzle_data['number_of_rooms'].append(rooms)

        # Price
        price_tag = listing.find('span', class_='ddc1b288')
        price = price_tag.text.strip() if price_tag else "N/A"
        dubizzle_data['price'].append(price)

        # Size
        area_tag = listing.find('span', attrs={'aria-label': 'Area'})
        area_span = area_tag.find('span', class_='_3e1113f0') if area_tag else None
        size = area_span.text.strip() if area_span else "N/A"
        dubizzle_data['size'].append(size)

    current_page += 1

    if current_page > MAX_PAGES:
        print("✅ Reached the maximum number of pages.")
        break

    time.sleep(1)  # Be polite to the server

# Optional: Export to CSV
df = pd.DataFrame(dubizzle_data)
df.to_csv("dubizzle_sale.csv", index=False)
print(f"\n✅ Scraping complete. {len(df)} listings saved to 'dubizzle_sale.csv'")
```
